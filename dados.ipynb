```python
import dask.dataframe as dd # import pandas as pd
import dask.array as da # import numpy as np
```

Abrindo dataframe usando Dask, com chunks de 25MB

```python
df = dd.read_csv('Acidentes_DadosAbertos_20230812.csv', blocksize=25e6, delimiter=';') # 25MB chunks  [manter suas partições com tamanho inferior a 100 MB]
```

```python
# Gerando estatísticas descritivas que resumem a tendência central 
print(df.describe().compute())
```

```python
# Verificando a quantidade de valores faltantes por coluna
df.isnull().sum().compute()
```

```python
# Filtrando apenas as colunas com valor nulo
null_counts = df.isnull().sum().compute()
print(null_counts[null_counts > 0])
```

Correlação das colunas numéricas

```python
# Filtrando apenas colunas numéricas
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
# Calculando correlação
corr_matrix = numeric_cols.corr().compute()  
corr_matrix
```
